```python
def factorial(n):
  if(n==1 or n==0):
    return 1
  else:
    return n*factorial(n-1)
n=int(input("enter the n value:"))
res=factorial(n)
print(res)
```

```python
def fibionacci(n):
  if n<=0:
    print("incorrect input")
  elif n==1:
    return 0
  elif n==2:
    return 1
  else:
    return fibionacci(n-1)+fibionacci(n-2)
n=int(input("ENTER NUMBER OF TERMS:"))
for i in range (1,n+1):
  print(fibionacci(i))
```
